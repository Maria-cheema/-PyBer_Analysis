```python
# Change the color of the markers to red.
plt.plot(y_axis, x_axis, 'o', c="red", label="Chicago")

# Add a legend for the city of Chicago.
plt.legend()

# Add a title and axes labels.
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")

# Add a limit for the x-axis data.
plt.xlim(0, 50)

# Invert the y-axis so that January is at the top
plt.gca().invert_yaxis()
```

```python
# Create a Bubble Chart Using the MATLAB Method
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# We can adjust the size by multiplying the data in the y-axis by any value.
# Let's multiply each data point in the y-axis by 3 and see what happens. 
# To do this, we can iterate through the y-axis data and multiply each data point by 3 and add it to a new list, like this:

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Using the list comprehension technique to multiply each data point in the y-axis list
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# To create a bubble chart using the object-oriented interface approach, we'll add the s parameter just like we did when we used the MATLAB approach.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

# 5.1.9 Chart Extras
### Add Error Bars to a Line Chart

```python
# Create a new Jupyter Notebook file
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# We can make the error bars look more professional by adding a "cap" to the error bars that's equal to a scalar value.
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# add error bars and a capsize using the object-oriented approach.
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

## Add Error Bars to a Bar Chart

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

## Create the same graph using the object-oriented approach.

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

## Add Minor Ticks


```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
